```python
# Importare la libreria yfinance e pandas
import yfinance as yf
import pandas as pd

# Crea un oggetto ticker per Tesla
tesla = yf.Ticker("TSLA")

# Estrai i dati storici delle azioni di Tesla
tesla_data = tesla.history(period="max")

# Reimposta l'indice
tesla_data.reset_index(inplace=True)

# Mostra le prime cinque righe del dataframe
tesla_data.head()
```

```python
# Importa le librerie necessarie
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Scarica la pagina web con i dati delle entrate di Tesla
url_tesla = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url_tesla).text

# Analizza il contenuto HTML usando BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')

# Estrai la tabella contenente i dati delle entrate di Tesla
tables = pd.read_html(str(soup))
tesla_revenue = tables[1]  # La seconda tabella contiene i dati delle entrate

# Rinominare le colonne per maggiore chiarezza
tesla_revenue.columns = ["Date", "Revenue"]

# Rimuovere i caratteri di formattazione ($ e ,) dalla colonna 'Revenue'
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(',|\$',"", regex=True)

# Rimuovere eventuali valori nulli o righe vuote
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

# Visualizza le ultime cinque righe del dataframe
tesla_revenue.tail()
```

```python
# Importare la libreria yfinance e pandas
import yfinance as yf
import pandas as pd

# Crea un oggetto ticker per GameStop
gme = yf.Ticker("GME")

# Estrai i dati storici delle azioni di GameStop
gme_data = gme.history(period="max")

# Reimposta l'indice
gme_data.reset_index(inplace=True)

# Mostra le prime cinque righe del dataframe
gme_data.head()
```

```python
# Importa le librerie necessarie
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Scarica la pagina web con i dati delle entrate di GameStop
url_gme = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data_2 = requests.get(url_gme).text

# Analizza il contenuto HTML usando BeautifulSoup
soup_2 = BeautifulSoup(html_data_2, 'html.parser')

# Estrai la tabella contenente i dati delle entrate di GameStop
tables_2 = pd.read_html(str(soup_2))
gme_revenue = tables_2[1]  # La seconda tabella contiene i dati delle entrate

# Rinominare le colonne per maggiore chiarezza
gme_revenue.columns = ["Date", "Revenue"]

# Rimuovere i caratteri di formattazione ($ e ,) dalla colonna 'Revenue'
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(',|\$',"", regex=True)

# Rimuovere eventuali valori nulli o righe vuote
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]

# Visualizza le ultime cinque righe del dataframe
gme_revenue.tail()



```

```python
# Definizione della funzione make_graph (se non già presente)
def make_graph(stock_data, revenue_data, stock):
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    import pandas as pd

    fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                        subplot_titles=("Historical Share Price", "Historical Revenue"),
                        vertical_spacing = .3)

    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']

    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date),
                             y=stock_data_specific.Close.astype("float"),
                             name="Share Price"), row=1, col=1)

    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date),
                             y=revenue_data_specific.Revenue.astype("float"),
                             name="Revenue"), row=2, col=1)

    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)

    fig.update_layout(showlegend=False, height=900, title=stock, xaxis_rangeslider_visible=True)
    fig.show()

# Esegui il grafico dei dati azionari di Tesla
make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
# Esegui il grafico dei dati azionari di GameStop
make_graph(gme_data, gme_revenue, 'GameStop')
```
